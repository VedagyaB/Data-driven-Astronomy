```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The file GW_data_file.csv contains the timeseries and the frequency data of the first Gravitational Wave detection made by LIGO, GW150914.
# The data file has 3 arrays:
# a) the strain data, which is filtered, so you don't have to worry about noises,
# b) the frequency of the data at that particular time (Not exactly),
# c) the time.

# Your first task is to make two plots:
# (a) strain vs time
# (b) frequency vs time.
```

```python
# Load data
data, freq, time = np.loadtxt('GW_data_file.csv',delimiter=',')

#Plotting begins
plt.figure(figsize=(15,3))
plt.plot(time,data, 'r')
plt.xlabel('Time(s)')
plt.ylabel('Normalized Strain') # Normalized Strain is not strain
plt.grid()
plt.show()
```

```python
plt.plot(time,freq, 'k')
plt.xlabel('Time(s)')
plt.ylabel('Frequency(Hz)')
plt.grid()
plt.show()
```

```python
zero_indices=[]

for i in range(len(data)-1):
    if data[i]==0 and data[i+1]!=0:
        zero_indices.append(i)
    elif (data[i]>0 and data[i+1]<0) or (data[i]<0 and data[i+1]>0):
        zero_indices.append(i)
    elif data[i]==0 and data[i+1]==0:
        zero_indices.append(i)
        i=i+1
plt.scatter(time[zero_indices],freq[zero_indices],s=10, color='g', marker='^')
plt.ylabel('Frequency(Hz)')
plt.grid()
plt.show()
```
